## UNSUPERVISED MACHINE LEARNING

#### PCA on cancer dataset

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# load the dataset
(X_cancer, y_cancer) = load_breast_cancer(return_X_y = True)
```

```python
# standardize the data before pca
# this centers the data (zero mean) unit s.d

X_scaled = StandardScaler().fit(X_cancer).transform(X_cancer)
pd.DataFrame(X_scaled).head()
```

```python
# dimensionality reduction with PCA
pc_cancer = PCA(n_components = 2).fit(X_scaled).transform(X_scaled)
pc_cancer
```

```python
pd.DataFrame(pc_cancer).head()
```

```python
# visualisation
colors = ["blue", "yellow"]
label_names = ["malign", "benign"]
plt.figure()
plt.scatter(pc_cancer[:,0], pc_cancer[:,1], c=[ colors[i] for i in y_cancer])
plt.show()
```
